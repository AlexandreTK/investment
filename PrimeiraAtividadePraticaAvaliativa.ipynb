```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import numpy as np

def get_html_page(url):
    page = urlopen(url)
    soup_page = BeautifulSoup(page, 'html.parser')
    return soup_page

def get_indicator_assets(html_soup_page):
    rgRows = html_soup_page.findAll('tr', attrs={'class':'rgRow'})
    rgAltRows = html_soup_page.findAll('tr', attrs={'class':'rgAltRow'})
    rgRows = rgRows+rgAltRows

    indicator_assets = []
    for rgRow in rgRows:
        rawData = rgRow.findAll('span')
        codigo = rawData[0].contents[0]
        acao   = rawData[1].contents[0]
        tipo   = rawData[2].contents[0]
        qtd_teorica = rawData[3].contents[0]
        participacao = rawData[4].contents[0]
        asset_data = [codigo, acao, tipo, qtd_teorica, participacao]
        indicator_assets.append(asset_data)
    return indicator_assets
```

```python
# Obter os dados do website www.fundamentus.com.br
def get_fundamentus_data(asset_name):
    url = 'https://www.fundamentus.com.br/detalhes.php?papel=' + asset_name

    #display(url)
    page = urlopen(url)
    soup_page = BeautifulSoup(page, 'html.parser')
    #display(soup_page)
    
    content = soup_page.find('div', attrs={'class':'conteudo'})
    tables = content.findAll('table')
    fundamentalist_analysis = tables[2]
    
    labels_array = []
    td_label = fundamentalist_analysis.findAll('td', attrs={'class':'label'})
    for label in td_label:
        label_content = label.find('span', attrs={'class':'txt'})
        if label_content:
            label_content = label_content.contents
            if (len(label_content) > 0):
                labels_array.append(label_content[0])
#     display(labels_array)
    
    data_array = []
    td_data = fundamentalist_analysis.findAll('td', attrs={'class':'data'})
    #display(td_all_data)
    for data in td_data:

        data_content_oscil = data.find('span', attrs={'class':'oscil'})
        if data_content_oscil:
            data_content_oscil = data_content_oscil.contents
            #display(data_content_oscil)
            if (len(data_content_oscil) > 0):
                data_content_oscil = data_content_oscil[0].contents[0]
                cleaned_content = data_content_oscil[0].split()[0]
                data_array.append(data_content_oscil[0])
                
        data_content_data = data.find('span', attrs={'class':'txt'})
        if data_content_data:
            data_content_data = data_content_data.contents
            #display(data_content)
            if (len(data_content_data) > 0):
                cleaned_content = data_content_data[0].split()[0]
                data_array.append(cleaned_content)
#     display(data_array)
    
    if (len(labels_array) != len(data_array)):
        print('ERROR PARSING')
    

    dictionary_label_data = dict(zip(labels_array, data_array))
    return dictionary_label_data
```

### índice Sharpe

Considerando "risk free rate" = 0.05, obtemos os seguintes resultados:

```python
def get_sharpe_index(close_price):
    asset_log_return = (np.log(close_price[1:]).values - np.log(close_price[0:-1]).values)
    asset_return = close_price[1:].values/close_price[0:-1].values - 1

    # Volatility
    anualized_volatility = asset_return.std() * 252**0.5
    #print('volatilidade anual:', anualized_volatility)


    # Sum of the log returns
    days = len(asset_return)
    total_return = asset_log_return.sum()
    anualized_return = (1+total_return)**(252/days) - 1
    #print('retorno anual:', anualized_return)

    sharpe = (anualized_return - 0.05) / anualized_volatility
    return sharpe
  
for asset in data_top5.columns.levels[1].values:
    print(asset)
    data = data_top5.xs(asset, axis=1, level=1)
    close_price = data['Close']
    
    sharpe = get_sharpe_index(close_price)
    print('sharpe:', sharpe)
    print()
```

### Fator Beta

Beta = Covariância entre o Retorno do Ativo e do Mercado / Variância do Retorno do Mercado

```python
def get_beta_index(asset_close_price, bov_close_price):
    
    # Calculating returns
    bov_return = bov_close_price[1:].values/close_price_bov[0:-1].values - 1
    asset_return = asset_close_price[1:].values/asset_close_price[0:-1].values - 1

    asset_bov_vector = np.vstack([asset_return, bov_return])
    # Removing nan's
    asset_bov_vector = np.ma.compress_cols(np.ma.masked_invalid(asset_bov_vector))

    # Computing covariance matrix
    covariance_asset_bov = np.cov(asset_bov_vector)
    beta = covariance_asset_bov[1,0] / covariance_asset_bov[1,1]

    return beta
  
for asset in stock_data.columns.levels[1].values[:-1]:
    print(asset)
    data_asset = stock_data.xs(asset, axis=1, level=1)
    data_bov = stock_data.xs('^BVSP', axis=1, level=1)
    close_price_asset = data_asset['Close']
    close_price_bov   = data_bov['Close']
    

    beta = get_beta_index(close_price_asset, close_price_bov)
    print('beta:', beta)
    print()
```

### Diagrama Risco x Retorno

```python
data_top5_prices = data_top5['Close']

data_top5_ptc = data_top5_prices.pct_change()
num_assets = len(data_top5_ptc.columns)

data_top5_ptc.iloc[0]= np.zeros(num_assets)


new_columns = ['RET_' + x for x in data_top5_ptc.columns]
data_top5_ptc.columns=[new_columns]
#display(data_top5_ptc)
```

```python
C   = data_top5_ptc.cov()
sig = data_top5_ptc.std()
MU  = data_top5_ptc.mean()

I = pd.DataFrame(np.identity(num_assets))
SIGMA = (C*(I.values))**0.5

P = data_top5_ptc.corr()
#ONE = [1,1,1,1]
ONE = np.ones(num_assets)
N=1000
ret=pd.DataFrame(np.zeros((N,1)))
risk=pd.DataFrame(np.zeros((N,1)))
MU
```

```python
for i in range(0, N):
    qabs = np.random.normal(0,100, num_assets)  
    qmx=qabs.max()
    qn=qabs/qmx
    qn1=qn/qn.sum()
    q=qn1.round(2)

    if i==1:
        q = [1/num_assets for i in range(num_assets)]
    ret[0][i]=np.dot(pd.DataFrame(q).transpose(),MU)
    risk[0][i]=(np.dot(np.dot(pd.DataFrame(q).transpose(),C),pd.DataFrame(q)))**0.5

fig = plt.subplots()
plt.plot(100*risk,100*ret,'r*')

plt.title('Teoria Moderna do Portfólio - Markowitz (1952),')
plt.ylabel('Retorno do Capital Mensal [%]')
plt.xlabel('Risco do Capital mensal [%]')
plt.show()
```

### Curva de Eficiência

```python
a=np.dot(pd.DataFrame(MU).transpose(),np.dot(np.linalg.inv(C),pd.DataFrame(MU)))
b=np.dot(pd.DataFrame(ONE).transpose(),np.dot(np.linalg.inv(C),pd.DataFrame(MU)))
c=np.dot(pd.DataFrame(MU).transpose(),np.dot(np.linalg.inv(C),pd.DataFrame(ONE)))
d=np.dot(pd.DataFrame(ONE).transpose(),np.dot(np.linalg.inv(C),pd.DataFrame(ONE)))
target_mus=np.arange(-0.1,0.1,0.0002)
risk_at_target_mus=np.zeros(len(target_mus))
for i in range(0, len(target_mus)):
    target_mu=target_mus[i]
    v1=np.linalg.det([[target_mu,b[0][0]],[1,d[0][0]]])*np.dot(np.linalg.inv(C),pd.DataFrame(MU))
    v2=np.linalg.det([[a[0][0],target_mu],[c[0][0],1]])*np.dot(np.linalg.inv(C),pd.DataFrame(ONE))
    v3=np.linalg.det([[a[0][0],b[0][0]],[c[0][0],d[0][0]]])
    opt_q=(v1+v2)/v3
    risk_at_target_mus[i]=(np.dot(np.dot(pd.DataFrame(opt_q).transpose(),C),pd.DataFrame(opt_q)))**0.5
fig = plt.subplots()
plt.plot(100*risk,100*ret,'r*')
plt.plot(100*risk_at_target_mus,100*target_mus,'b*')

for asset in range(num_assets):
    asset_name = 'A'+str(asset+1)
    plt.plot(100*sig[asset], 100*MU[asset], 'g*', markersize=15.0)  
    plt.text(100*sig[asset]+0.1, 100*MU[asset], asset_name, fontsize=16)

#plt.plot(100*risk_minrisk,100*mu_minrisk,'k*',markersize=15.0)
plt.axis([-0.5,12,-0.5,3])
plt.title('Teoria Moderna do Portfólio - Markowitz (1952),')
plt.ylabel('Retorno do Capital Mensal [%]')
plt.xlabel('Risco do Capital mensal [%]')
plt.show()
```
